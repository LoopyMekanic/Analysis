```python
import numpy as np
```

```python
# Slicing NumPy Arrays - Return 2,3,4,5| Slice 1:5 returns 1 up till 5 but not including 5
np1 = np.array(["Tony", "Tim", "Mary", "Bob", "Pete", "April", "Wes", "Steve", "Aspen", "Ivy"])
print(np1[1:5])
np2 = np1[1:5]
print(np2)
```

```python
# Return from something till end of array
print(np1[3:]) # : nothing means till the end
```

```python
# Return negative slices, from the end backwards
print(np1[-3:-1])
```
